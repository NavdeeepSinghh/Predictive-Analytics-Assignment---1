```python
import numpy as np
import pandas as pd
import math

ROLL_NO = 102303787
CSV_PATH = "/content/data3.csv"

df = pd.read_csv(CSV_PATH, encoding="latin1")
df.columns = df.columns.astype(str).str.strip()

NO2_COL = "no2"

x = pd.to_numeric(df[NO2_COL], errors="coerce").dropna()
x = x[x >= 0].to_numpy(dtype=float)

a_r = 0.05 * (ROLL_NO % 7)
b_r = 0.3 * ((ROLL_NO % 5) + 1)

t = np.clip(b_r * x, -1.0, 1.0)
z = x + a_r * np.arcsin(t)

mu = float(np.mean(z))
sigma2 = float(np.mean((z - mu) ** 2))
lam = float(1.0 / (2.0 * sigma2))
c = float(math.sqrt(lam / math.pi))

print("Lambda (λ) :", lam)
print("Mu (μ)     :", mu)
print("c          :", c)
```
